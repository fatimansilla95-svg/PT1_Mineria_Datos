```python
import pandas as pd
df_original = pd.read_json("/content/sample_data/streaming_users_dirty.json")
```

```python
# Inicializar etl_log si no está definido (idealmente en una celda anterior para persistencia)
if 'etl_log' not in locals() and 'etl_log' not in globals():
    etl_log = []

def registrar_log_etl(paso, descripcion, df_actual):
    log_entry = {
        "paso": paso,
        "descripcion": descripcion,
        "num_filas": len(df_actual),
        "num_columnas": len(df_actual.columns),
        "columnas": list(df_actual.columns)
    }
    etl_log.append(log_entry)
    print(f"Log registrado para el paso: {paso}")

registrar_log_etl(
    paso="01_Carga_Inicial",
    descripcion="Lectura del dataset crudo desde la carpeta raw.",
    df_actual=df_original
)
```

```python
df_sin_duplicados = df_original.drop_duplicates()

registrar_log_etl(
    paso="02_Remocion_Duplicados",
    descripcion="Se eliminan filas idénticas repetidas.",
    df_actual=df_sin_duplicados
)
```

```python
import os

df_limpio = df_sin_duplicados.dropna(subset=["user_id"])

# Crear el directorio si no existe
os.makedirs("data/processed", exist_ok=True)

df_limpio.to_csv("data/processed/dataset_limpio.csv", index=False)

registrar_log_etl(
    paso="03_Limpieza_Nulos_Final",
    descripcion="Filas con nulos en variables clave eliminadas. Dataset listo.",
    df_actual=df_limpio
)
```
