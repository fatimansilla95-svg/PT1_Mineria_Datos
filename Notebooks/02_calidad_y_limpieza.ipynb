**Preparación de datos**

Selección de columnas numéricas

```python
import pandas as pd
df = pd.read_json("/content/sample_data/streaming_users_dirty.json")
num_cols = df.select_dtypes(include="number").columns
```

**Limpiar y transformar**

```python
# Eliminar filas duplicadas
df = df.drop_duplicates()
```

```python
# Rellenar valores faltantes en columnas numéricas con la mediana
num_cols = df.select_dtypes(include="number").columns
for col in num_cols:
    df[col] = df[col].fillna(df[col].median())

# Rellenar valores faltantes en columnas categóricas con el valor más frecuente
cat_cols = df.select_dtypes(include="object").columns
for col in cat_cols:
    df[col] = df[col].fillna(df[col].mode()[0])
```

Eliminamos filas duplicadas
**Usamos la mediana porque es más robusta frente a valores extremos**
